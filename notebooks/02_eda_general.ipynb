```python
import sys
from pathlib import Path

BASE_DIR = Path(r"D:\Python\spotify_rock")  # ajusta si tu ruta es otra
SRC_DIR = BASE_DIR / "src"
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
sns.set()

from data.spotify_api import get_spotify_client, fetch_playlist_tracks_with_features

sp = get_spotify_client()
```

```python
playlist_url = "https://open.spotify.com/playlist/PEGAR_AQUI_URL"
playlist_id = playlist_url.split("/")[-1].split("?")[0]
playlist_id
```

```python
df_pl = fetch_playlist_tracks_with_features(sp, playlist_id)
df_pl.shape, df_pl.head()
```

```python
df_ranked = df_pl.sort_values("popularity", ascending=False).reset_index(drop=True)
df_ranked.head(20)
```

```python
top_n = 15
top_tracks = df_ranked.head(top_n)

plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_tracks,
    x="popularity",
    y="track_name",
    hue="artist_name",
    dodge=False,
    legend=False,
)
plt.title(f"Top {top_n} canciones por popularidad en la playlist")
plt.xlabel("Popularidad (0-100)")
plt.ylabel("Canción")
plt.show()
```
